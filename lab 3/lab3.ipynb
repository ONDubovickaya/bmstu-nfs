# Лабораторная №3
### Набор данных -- Научные тексты разных стилей (12 штук)

```python
import pandas as pd
import numpy as np

from nltk.corpus import stopwords
from wordcloud import WordCloud

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score

import matplotlib.pyplot as plt
import torch

from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

from sklearn.metrics import adjusted_rand_score
from sklearn.preprocessing import LabelEncoder

from transformers import AutoTokenizer, AutoModel

import re
import nltk
from nltk.stem.snowball import SnowballStemmer
```

### Чтение данных

```python
data = pd.read_csv('data.csv')
data
```

### WordCloud

```python
# определение стоп-слов на русском языке
stop_words = stopwords.words('russian') 
stopwords = set(stop_words)

comment_words = ''

for val in data['text']:
    val = str(val) 
    # разбиение строки на слова по пробелам
    tokens = val.split() 
    
    # перевод всех слов строки в нижний регистр
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 
      
    comment_words += " ".join(tokens)+" "

# создание облака слов
wordcloud = WordCloud(width = 2048, height = 2048, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 10).generate(comment_words) 
                
plt.figure(figsize = (20, 20), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show()
```

### 1) [БЕЗ ПРЕДОБРАБОТКИ]
### TF-IDF

```python
data_tf = data['text']

# создание объекта TF-IDF векторизатора
tfidf_vectorizer = TfidfVectorizer(max_df=0.85, max_features=1000)

# преобразование текстов в TF-IDF векторы
tfidf_matrix = tfidf_vectorizer.fit_transform(data_tf)

# для разного кол-ва кластеров оценим качество кластеризации
min_clusters = 2
max_clusters = 10
silhouette_scores = []

for num_clusters in range(min_clusters, max_clusters + 1):
    clustering_model = AgglomerativeClustering(n_clusters=num_clusters)
    cluster_labels = clustering_model.fit_predict(tfidf_matrix.toarray())
    silhouette_avg = silhouette_score(tfidf_matrix, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# визуализация р-тов методом "локтя"
plt.plot(range(min_clusters, max_clusters + 1), silhouette_scores, marker='o')
plt.title("Метод 'локтя' для определения числа кластеров")
plt.xlabel("Кол-во кластеров")
plt.ylabel("Средний коэффициент силуэта")
plt.show()

# выбор оптимального кол-ва кластеров
optimal_num_clusters = silhouette_scores.index(max(silhouette_scores)) + min_clusters
print("Оптимальное кол-во кластеров (TF-IDF):", optimal_num_clusters)
```

```python
# создание объекта иерархической кластеризации с оптимальным числом кластеров
clustering_model = AgglomerativeClustering(n_clusters=optimal_num_clusters)

# применение кластеризации к TF-IDF векторам
cluster_labels_tf = clustering_model.fit_predict(tfidf_matrix.toarray())

# добавление метки кластеров к данным
clustered_data_tf = pd.DataFrame({"text": data_tf, "cluster_tf": cluster_labels_tf})

# вывод р-тов кластеризации
print(clustered_data_tf)
```

### BERT (контекстуализированная векторная модель)

```python
data_bert = data['text']

# инициализация токенизатора и модели BERT
tokenizer = AutoTokenizer.from_pretrained("cointegrated/rubert-tiny")
model = AutoModel.from_pretrained("cointegrated/rubert-tiny")

# получение векторов текстов
embeddings = []

for description in data_bert:
    inputs = tokenizer(description, return_tensors='pt', padding=True, truncation=True)
    outputs = model(**inputs)
    embeddings.append(outputs.last_hidden_state.mean(dim=1).squeeze().detach())

embeddings = torch.stack(embeddings)

# для разного кол-ва кластеров оценим качество кластеризации
min_clusters = 2
max_clusters = 10
silhouette_scores = []

for num_clusters in range(min_clusters, max_clusters + 1):
    clustering_model = AgglomerativeClustering(n_clusters=num_clusters)
    cluster_labels = clustering_model.fit_predict(embeddings.cpu().numpy())
    silhouette_avg = silhouette_score(embeddings.cpu().numpy(), cluster_labels)
    silhouette_scores.append(silhouette_avg)
    
# визуализация р-тов методом "локтя"
plt.plot(range(min_clusters, max_clusters + 1), silhouette_scores, marker='o')
plt.title("Метод 'локтя' для определения числа кластеров")
plt.xlabel("Кол-во кластеров")
plt.ylabel("Средний коэффициент силуэта")
plt.show()

# выбор оптимального кол-ва кластеров
optimal_num_clusters = silhouette_scores.index(max(silhouette_scores)) + min_clusters
print("Оптимальное кол-во кластеров (BERT):", optimal_num_clusters)
```

```python
# создание объекта иерархической кластеризации с оптимальным числом кластеров
clustering_model = AgglomerativeClustering(n_clusters=optimal_num_clusters)

# применение кластеризации к векторам эмбеддингов
cluster_labels_bert = clustering_model.fit_predict(embeddings.cpu().numpy())

# добавление метки кластеров к данным
clustered_data_bert = pd.DataFrame({"text": data_bert, "cluster_bert": cluster_labels_bert})

# вывод р-тов кластеризации
print(clustered_data_bert)
```

### Визуализация кластеров (PCA)

```python
# применение метода главных компонент (PCA) для уменьшения размерности до 2D
pca = PCA(n_components=2)
tfidf_embeddings_2d = pca.fit_transform(tfidf_matrix.toarray())
embeddings_2d = pca.fit_transform(embeddings.cpu().numpy())

fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# график для TF-IDF 
for cluster_num in range(np.max(cluster_labels_tf) + 1):
    cluster_mask = (cluster_labels_tf == cluster_num)
    axes[0].scatter(tfidf_embeddings_2d[cluster_mask, 0], tfidf_embeddings_2d[cluster_mask, 1], label=f'Кластер {cluster_num + 1}')
axes[0].set_title('TF-IDF')
axes[0].legend()

# график для BERT
for cluster_num in range(np.max(cluster_labels_bert) + 1):
    cluster_mask = (cluster_labels_bert == cluster_num)
    axes[1].scatter(embeddings_2d[cluster_mask, 0], embeddings_2d[cluster_mask, 1], label=f'Кластер {cluster_num + 1}')
axes[1].set_title('BERT')
axes[1].legend()

plt.tight_layout()
plt.show()
```

### Объединение результатов кластеризаций по TF-IDF и BERT

```python
# объединение р-тов двух кластеризаций
merged_data = pd.merge(clustered_data_bert, clustered_data_tf, on='text', suffixes=('_bert', '_tf'))

# признаки объединённой кластеризации
features = merged_data[['cluster_bert', 'cluster_tf']].values

# для разного кол-ва кластеров оценим качество кластеризации
min_clusters = 2
max_clusters = 4
silhouette_scores = []

for num_clusters in range(min_clusters, max_clusters + 1):
    clustering_model = KMeans(n_clusters=num_clusters)
    cluster_labels = clustering_model.fit_predict(features)
    silhouette_avg = silhouette_score(features, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# визуализация р-тов методом "локтя"
plt.plot(range(min_clusters, max_clusters + 1), silhouette_scores, marker='o')
plt.title("Метод 'локтя' для определения числа кластеров (второй уровень)")
plt.xlabel("Кол-во кластеров")
plt.ylabel("Средний коэффициент силуэта")
plt.show()

# выбор оптимального кол-ва кластеров
optimal_num_clusters = silhouette_scores.index(max(silhouette_scores)) + min_clusters
print("Оптимальное кол-во кластеров (второй уровень):", optimal_num_clusters)
```

```python
# создание объекта KMeans с оптимальным числом кластеров
clustering_model = KMeans(n_clusters=optimal_num_clusters)

# применение кластеризацию ко вторым уровням кластеров
second_level_clusters = clustering_model.fit_predict(features)

# добавление метки второго уровня кластеров к объединённым данным
merged_data['second_level_cluster'] = second_level_clusters

# вывод р-тов второй кластеризации
merged_data
```

### Оценка сходства с исходными метками

```python
# преобразование текстовых меток стилей в числовое представление
label_encoder = LabelEncoder()
true_style_labels_numeric = label_encoder.fit_transform(data['style'])

# извлечение предсказанных меток кластеров из финальной модели
predicted_labels = merged_data['second_level_cluster']

# вычисление ARI для оценки сходства с исходными метками
ari_style = adjusted_rand_score(true_style_labels_numeric, predicted_labels)

print("ARI для стилей (БЕЗ ПРЕДОБРАБОТКИ):", round(ari_style, 3))
```

### 2) [С ПРЕДОБРАБОТКОЙ]
### Предобработка

```python
nltk.download('averaged_perceptron_tagger_ru')

# ф-ция, которая очищает текст от всех символов, кроме кириллического и базового латинского алфавита, стоп-слов и функциональных частей речи
# она возвращает строку из основ слов
def clean_text(string):
    # общая очистка текста
    string = string.lower()
    string = re.sub(r"http\S+", "", string)
    string = str.replace(string,'Ё','е')
    string = str.replace(string,'ё','е')
    prog = re.compile('[А-Яа-яA-Za-z]+')
    words = prog.findall(string.lower())
    
    # удаление стоп-слов
    stopwords = nltk.corpus.stopwords.words('russian')
    words = [w for w in words if w not in stopwords]
    
    # удаление слов с функциональными частями речи
    functional_pos = {'CONJ', 'PRCL'}
    words = [w for w, pos in nltk.pos_tag(words, lang='rus') if pos not in functional_pos]
    
    # приведение слов к словоформе
    stemmer = SnowballStemmer('russian')
    return ' '.join(list(map(stemmer.stem, words)))

# обновление столбца 'text' в новом датафрейме
df_preprocess = data
df_preprocess['text'] = list(map(clean_text, data['text']))

df_preprocess
```

### TF-IDF

```python
data_tf = df_preprocess['text']

# создание объекта TF-IDF векторизатора
tfidf_vectorizer = TfidfVectorizer(max_df=0.85, max_features=1000)

# преобразование текстов в TF-IDF векторы
tfidf_matrix = tfidf_vectorizer.fit_transform(data_tf)

# для разного кол-ва кластеров оценим качество кластеризации
min_clusters = 2
max_clusters = 5
silhouette_scores = []

for num_clusters in range(min_clusters, max_clusters + 1):
    clustering_model = AgglomerativeClustering(n_clusters=num_clusters)
    cluster_labels = clustering_model.fit_predict(tfidf_matrix.toarray())
    silhouette_avg = silhouette_score(tfidf_matrix, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# визуализация р-тов методом "локтя"
plt.plot(range(min_clusters, max_clusters + 1), silhouette_scores, marker='o')
plt.title("Метод 'локтя' для определения числа кластеров")
plt.xlabel("Кол-во кластеров")
plt.ylabel("Средний коэффициент силуэта")
plt.show()

# выбор оптимального кол-ва кластеров
optimal_num_clusters = silhouette_scores.index(max(silhouette_scores)) + min_clusters
print("Оптимальное кол-во кластеров (TF-IDF):", optimal_num_clusters)
```

```python
# создание объекта иерархической кластеризации с оптимальным числом кластеров
clustering_model = AgglomerativeClustering(n_clusters=optimal_num_clusters)

# применение кластеризации к TF-IDF векторам
cluster_labels_tf = clustering_model.fit_predict(tfidf_matrix.toarray())

# добавление метки кластеров к данным
clustered_data_tf = pd.DataFrame({"text": data_tf, "cluster_tf": cluster_labels_tf})

# вывод р-тов кластеризации
print(clustered_data_tf)
```

### BERT (контекстуализированная векторная модель)

```python
data_bert = df_preprocess['text']

# инициализация токенизатора и модели BERT
tokenizer = AutoTokenizer.from_pretrained("cointegrated/rubert-tiny")
model = AutoModel.from_pretrained("cointegrated/rubert-tiny")

# получение векторов текстов
embeddings = []

for description in data_bert:
    inputs = tokenizer(description, return_tensors='pt', padding=True, truncation=True)
    outputs = model(**inputs)
    embeddings.append(outputs.last_hidden_state.mean(dim=1).squeeze().detach())

embeddings = torch.stack(embeddings)

# для разного кол-ва кластеров оценим качество кластеризации
min_clusters = 2
max_clusters = 5
silhouette_scores = []

for num_clusters in range(min_clusters, max_clusters + 1):
    clustering_model = AgglomerativeClustering(n_clusters=num_clusters)
    cluster_labels = clustering_model.fit_predict(embeddings.cpu().numpy())
    silhouette_avg = silhouette_score(embeddings.cpu().numpy(), cluster_labels)
    silhouette_scores.append(silhouette_avg)
    
# визуализация р-тов методом "локтя"
plt.plot(range(min_clusters, max_clusters + 1), silhouette_scores, marker='o')
plt.title("Метод 'локтя' для определения числа кластеров")
plt.xlabel("Кол-во кластеров")
plt.ylabel("Средний коэффициент силуэта")
plt.show()

# выбор оптимального кол-ва кластеров
optimal_num_clusters = silhouette_scores.index(max(silhouette_scores)) + min_clusters
print("Оптимальное кол-во кластеров (BERT):", optimal_num_clusters)
```

```python
# создание объекта иерархической кластеризации с оптимальным числом кластеров
clustering_model = AgglomerativeClustering(n_clusters=optimal_num_clusters)

# применение кластеризации к векторам эмбеддингов
cluster_labels_bert = clustering_model.fit_predict(embeddings.cpu().numpy())

# добавление метки кластеров к данным
clustered_data_bert = pd.DataFrame({"text": data_bert, "cluster_bert": cluster_labels_bert})

# вывод р-тов кластеризации
print(clustered_data_bert)
```

### Визуализация кластеров (PCA)

```python
# применение метода главных компонент (PCA) для уменьшения размерности до 2D
pca = PCA(n_components=2)
tfidf_embeddings_2d = pca.fit_transform(tfidf_matrix.toarray())
embeddings_2d = pca.fit_transform(embeddings.cpu().numpy())

fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# график для TF-IDF 
for cluster_num in range(np.max(cluster_labels_tf) + 1):
    cluster_mask = (cluster_labels_tf == cluster_num)
    axes[0].scatter(tfidf_embeddings_2d[cluster_mask, 0], tfidf_embeddings_2d[cluster_mask, 1], label=f'Кластер {cluster_num + 1}')
axes[0].set_title('TF-IDF')
axes[0].legend()

# график для BERT
for cluster_num in range(np.max(cluster_labels_bert) + 1):
    cluster_mask = (cluster_labels_bert == cluster_num)
    axes[1].scatter(embeddings_2d[cluster_mask, 0], embeddings_2d[cluster_mask, 1], label=f'Кластер {cluster_num + 1}')
axes[1].set_title('BERT')
axes[1].legend()

plt.tight_layout()
plt.show()
```
